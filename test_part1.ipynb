```python
import cv2
import matplotlib.pyplot as plt

# Load the image
image_path = "C:\\Users\\grohi\\Seed project dp\\test_set2\\Worst\\Worst (56).JPG"
img = cv2.imread(image_path)

# Check if the image was loaded successfully
if img is None:
    print("Error: Unable to load image")
else:
    # Convert BGR to RGB color space
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Display the image
    plt.imshow(img_rgb)
    plt.title("Test image")
    plt.show()

```

```python
#test model
import tensorflow as tf

# Load the image
image = tf.keras.preprocessing.image.load_img(image_path, target_size=(128, 128))

# Convert image to numpy array
input_arr = tf.keras.preprocessing.image.img_to_array(image)
input_arr=np.array([input_arr])#convert single image to batch
print(input_arr.shape)
```

```python
predicition=model.predict(input_arr)
predicition,predicition.shape
```

```python
result_index=np.argmax(predicition)
result_index
```

```python
class_name=['Average', 'Bad', 'Excellent', 'Good', 'Worst']
```

```python
#output
model_prediction=class_name[result_index]
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title(f"the High-quality seeds of seed is:{model_prediction}")
plt.xticks([])
plt.yticks([])
plt.show()
```
